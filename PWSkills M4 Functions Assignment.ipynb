```python
# Q1) Sum of all even numbers in a list
def sum_of_evens(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(nums))
```

```python
#Q2) Reverse a string

def reverse_string(s):
    return s[::-1]

text = "hello"
print(reverse_string(text)) 
```

```python
#Q3) Square of each number in a list
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

nums = [1, 2, 3, 4]
print(square_numbers(nums))
```

```python
#Q4) Check if a number is prime (1 to 200)
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [num for num in range(1, 201) if is_prime(num)]
print(primes) 
```

```python
# Q5) Iterator class for Fibonacci sequence
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return result

fib = Fibonacci(10)
print(list(fib))
```

```python
# Q6) Generator for powers of 2
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

print(list(powers_of_two(5)))
```

```python
#Q7) Generator to read file line by line

def read_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
# Q8) Sort tuples based on the second element
tuples = [(1, 3), (4, 1), (2, 2)]
sorted_tuples = sorted(tuples, key=lambda x: x[1])
print(sorted_tuples) 
```

```python
# Q9) Convert Celsius to Fahrenheit using map()
celsius = [0, 20, 30, 40]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# Q10) Remove vowels from a string using filter()
def remove_vowels(string):
    return ''.join(filter(lambda ch: ch.lower() not in "aeiou", string))

text = "hello world"
print(remove_vowels(text))
```
